```python


import numpy as np
import matplotlib.pyplot as plt

# Define the Izhikevich neuron model parameters
a = 0.02
b = 0.2
c = -65.0
d = 8.0

# Define the simulation parameters
t_start = 0.0
t_end = 1000.0
dt = 0.01
v_init = c
u_init = b * v_init
T = 100.0 # period of the square wave

# Define the square wave function
def square_wave(t, T):
    return np.sign(np.sin(2.0*np.pi*t/T))

# Define the Izhikevich neuron function
def izhikevich_neuron(v, u, I):
    v_next = v + dt*(0.04*v**2 + 5*v + 140 - u + I)
    u_next = u + dt*a*(b*v - u)
    if v_next >= 30.0:
        v_next = c
        u_next = u + d
    return v_next, u_next

# Initialize the simulation
t = t_start
v = v_init
u = u_init
v_list = [v]
u_list = [u]
I_list = [square_wave(t, T)]

# Run the simulation
while t < t_end:
    t += dt
    v, u = izhikevich_neuron(v, u, square_wave(t, T))
    v_list.append(v)
    u_list.append(u)
    I_list.append(square_wave(t, T))

# Plot the results
plt.figure(figsize=(10,5))
plt.subplot(3,1,1)
plt.plot(np.arange(len(v_list))*dt, v_list)
plt.ylabel('Membrane\npotential (mV)', fontsize=12)
plt.subplot(3,1,2)
plt.plot(np.arange(len(u_list))*dt, u_list)
plt.ylabel('Recovery\nvariable (mV)', fontsize=12)
plt.subplot(3,1,3)
plt.plot(np.arange(len(I_list))*dt, I_list)
plt.xlabel('Time (ms)', fontsize=12)
plt.ylabel('Input\ncurrent (nA)', fontsize=12)
plt.tight_layout()
plt.show()
```
