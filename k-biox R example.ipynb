```r
# Load required libraries
library(MASS) # for generating multivariate normal distributions
library(ggplot2) # for plotting
library(pracma) # for PCA

# Set seed for reproducibility
set.seed(123)

# Generate data
mean1 <- c(0, 0)
mean2 <- c(10, 10)
cov1 <- diag(2^2, 2) # covariance matrix for first distribution
cov2 <- diag(3^2, 2) # covariance matrix for second distribution

# Generate Gaussian distributions
data1 <- mvrnorm(n = 100, mu = mean1, Sigma = cov1)
data2 <- mvrnorm(n = 100, mu = mean2, Sigma = cov2)

# Plot the data
ggplot() +
  geom_point(aes(x = data1[,1], y = data1[,2]), color = "blue", alpha = 0.5) +
  geom_point(aes(x = data2[,1], y = data2[,2]), color = "red", alpha = 0.5) +
  labs(x = "X", y = "Y", title = "2D Gaussian Distributions") +
  theme_minimal()

# Combine the data
data <- rbind(data1, data2)

# Perform PCA
pca_result <- prcomp(data, scale. = TRUE)

# Plot the projected data points
projected_data <- data %*% pca_result$rotation
ggplot() +
  geom_point(aes(x = projected_data[,1], y = projected_data[,2]), color = "green", alpha = 0.5) +
  labs(x = "PC1", y = "PC2", title = "Projected Data Points after PCA") +
  theme_minimal()

# Print out eigenvalues and eigenvectors
print("Eigenvalues:")
print(pca_result$sdev^2)
print("Eigenvectors:")
print(pca_result$rotation)
```
