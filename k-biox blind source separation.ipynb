```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from sklearn.decomposition import FastICA

# Generate the original sources
t = np.linspace(0, 1, 1000)
s1 = np.sin(2 * np.pi * 5 * t)
s2 = signal.sawtooth(2 * np.pi * 3 * t)

# Plot the original sources
plt.subplot(2, 1, 1)
plt.plot(t, s1)
plt.title('Original Sources')
plt.ylabel('Amplitude')
plt.subplot(2, 1, 2)
plt.plot(t, s2)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Generate the mixtures
A = np.array([[1, 0.5], [0.8, 1]])
X = np.dot(A, np.array([s1, s2]))

# Plot the mixtures
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(t, X[0])
plt.title('Mixtures')
plt.ylabel('Amplitude')
plt.subplot(2, 1, 2)
plt.plot(t, X[1])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Run ICA to recover the original sources
ica = FastICA(n_components=2)
S = ica.fit_transform(X.T)

# Plot the recovered sources
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(t, S[:, 0])
plt.title('Recovered Sources')
plt.ylabel('Amplitude')
plt.subplot(2, 1, 2)
plt.plot(t, S[:, 1])
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.show()
```
